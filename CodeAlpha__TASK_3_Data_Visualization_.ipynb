```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('housing.csv')
df.head()
```

```python
# scatter plot
plt.scatter(df['bathrooms'],df['price'])
plt.xlabel('bathrooms')
plt.ylabel('price')
plt.show()
```

```python
plt.scatter(df['bedrooms'],df['price'],c=df['view'],s=df['grade'])
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.colorbar()
plt.show()
```

```python
# line chart
plt.plot(df['sqft_living'])
plt.plot(df['sqft_lot'])
plt.xlabel('date')
plt.ylabel('sqft')
plt.title('sqft_living vs sqft_lot')
plt.legend(['sqft_living','sqft_lot'])
plt.show()
```

```python
# bar chart
plt.bar(df['floors'],df['price'])
plt.title('floors vs bedrooms')
plt.xlabel('floors')
plt.ylabel('price')
plt.show()
```

```python
# histogram
plt.hist(df['sqft_living'],bins=10)
plt.title('histogram of sqft_living')
plt.show()
```

```python
#line plot
sns.lineplot(x='date',y='price',data=df)
plt.show()
```

```python
sns.scatterplot(x='bathrooms',y='price',data=df)
plt.show()
```

```python
sns.scatterplot(x='sqft_living',y='price',data=df,hue='bedrooms')
plt.show()
```

```python
sns.barplot(x='floors',y='price',data=df)
plt.show()
```

```python
# histogram
sns.histplot(df['sqft_living'],bins=10)
plt.title('histogram of sqft_living')
plt.show()
```
